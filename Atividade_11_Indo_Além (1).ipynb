```python
from transformers import pipeline

# 1. Definir o T√≥pico e a Tarefa
print("## üè∑Ô∏è Aplica√ß√£o de Demonstra√ß√£o: Classifica√ß√£o de Sentimentos")
print("---")

# 2. Carregar o Pipeline
# Usaremos um classificador de sentimentos pr√©-treinado e eficiente.
# Este √© um exemplo de modelo de Classifica√ß√£o de Sentimento (LLM/Transformer menor)
# que faz parte do ecossistema HuggingFace.
try:
    classifier = pipeline(
        "sentiment-analysis",
        model="distilbert-base-uncased-finetuned-sst-2-english"
    )
    print("Modelo de Classifica√ß√£o de Sentimentos carregado com sucesso.")

    # 3. Textos de Exemplo
    textos = [
        "A biblioteca HuggingFace √© extremamente √∫til e bem documentada para LLMs.",
        "Meu c√≥digo apresentou um erro estranho e estou muito frustrado com o resultado.",
        "Apesar do contratempo, o novo LLM me ajudou a refatorar o c√≥digo rapidamente."
    ]

    print("\n## üìù Resultados da Classifica√ß√£o:")

    # 4. Processar e Exibir Resultados
    for texto in textos:
        resultado = classifier(texto)
        sentimento = resultado[0]['label']
        confianca = resultado[0]['score'] * 100

        # Formata a sa√≠da
        print(f"\n---")
        print(f"Texto: **\"{texto}\"**")
        print(f"Sentimento: **{sentimento}**")
        print(f"Confian√ßa: **{confianca:.2f}%**")

except Exception as e:
    print(f"\nOcorreu um erro ao carregar o pipeline: {e}")
    print("Verifique sua conex√£o com a internet ou se as bibliotecas est√£o instaladas corretamente.")
```
