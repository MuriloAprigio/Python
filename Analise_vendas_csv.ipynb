```python
# 01 Percorrer todos os arquivos da pasta base de dados (Pasta Vendas)

import os
import pandas as pd
import plotly.express as px

lista_arquivo = os.listdir("/content/drive/MyDrive/Python/Vendas")
display(lista_arquivo)


```

```python
tabela_total = pd.DataFrame()

# 02 Importar base de dados de Vendas

#For--variável--lista de item


for arquivo in lista_arquivo:
#se tem 'Vendas' no nome do arquivo então:
  if "Vendas" in arquivo:
 #Importar o arquivo
   tabela = pd.read_csv(f'/content/drive/MyDrive/Python/Vendas/{arquivo}')
   tabela_total = tabela_total.append(tabela)
   # 03 tratar/ compilar base dados
display(tabela_total)
```

```python
#Calcular o produto mais vendido (em quantidade)

tabela_produto = tabela_total.groupby('Produto').sum()
tabela_produto = tabela_produto[['Quantidade Vendida']].sort_values(by='Quantidade Vendida', ascending=False)
display(tabela_produto)
```

```python
#Calcular o produto que mais faturou
tabela_total['Faturamento'] = tabela_total['Quantidade Vendida'] * tabela_total['Preco Unitario']

tabela_faturamento = tabela_total.groupby('Produto').sum()
tabela_faturamento = tabela_faturamento[['Faturamento']].sort_values(by='Faturamento', ascending=False)
display(tabela_faturamento)
```

```python
#Calcular loja que mais vendeu

tabela_loja = tabela_total.groupby('Loja').sum()
tabela_loja = tabela_loja[['Faturamento']]
display(tabela_loja)
```

```python
#Criação de Dashboard

grafico =px.bar(tabela_loja,x=tabela_loja.index ,y='Faturamento' )
grafico.show()
```
